```python
answer = []
for number in range(1, 101):
  if number % 15 == 0:
    answer.append("FizzBuzz")
  elif number % 3 == 0:
    answer.append("Fizz")
  elif number % 5 == 0:
    answer.append("Buzz")
  else:
    answer.append(number)
print(answer)
```

```python
def generate_fizzbuzz(max_num):
  answer = []
  for number in range(1, max_num + 1):
    if number % 15 == 0:
      answer.append("FizzBuzz")
    elif number % 3 == 0:
      answer.append("Fizz")
    elif number % 5 == 0:
      answer.append("Buzz")
    else:
      answer.append(number)
  return answer

print(generate_fizzbuzz(100))
```
